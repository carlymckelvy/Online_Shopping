```python
import pandas as pd
import numpy as np
```

```python
file = 'training_sample.csv'
```

```python
# The correct encoding must be used to read the CSV in pandas
df = pd.read_csv(file, encoding="ISO-8859-1")
```

```python
df.head()
```

```python
df.count()
```

```python
# Drop all rows with missing information
df = df.dropna(how='any')
```

```python
# Verify dropped rows
df.count()
```

```python
# The Amount column is the wrong data type. It should be numeric.
df.dtypes
```

```python
compare_df = df[['UserID','returning_user','device_mobile','device_computer','device_tablet','ordered']]
compare_df
```

```python
conditions = [
#     all
    ((compare_df['device_mobile'] == 1) & (compare_df['device_computer'] == 1) & (compare_df['device_tablet'] == 1)),
    
#     mobile only; tablet only; computer only
    ((compare_df['device_mobile'] == 1) & (compare_df['device_computer'] == 0) & (compare_df['device_tablet'] == 0)),
    ((compare_df['device_mobile'] == 0) & (compare_df['device_computer'] == 0) & (compare_df['device_tablet'] == 1)),
    ((compare_df['device_mobile'] == 0) & (compare_df['device_computer'] == 1) & (compare_df['device_tablet'] == 0)),
    
#     mobile & computer;  mobile & tablet; computer & tablet;
    ((compare_df['device_mobile'] == 1) & (compare_df['device_computer'] == 1) & (compare_df['device_tablet'] == 0)),
    ((compare_df['device_mobile'] == 1) & (compare_df['device_computer'] == 0) & (compare_df['device_tablet'] == 1)),
    ((compare_df['device_mobile'] == 0) & (compare_df['device_computer'] == 1) & (compare_df['device_tablet'] == 1))
    ]
```

```python
values = ['All', 'Mobile Only', 'Tablet Only', 'Computer Only','Mobile & Computer','Mobile & Tablet','Computer & Tablet']
```

```python
compare_df['Device Type'] = np.select(conditions,values)
```

```python
compare_df['returning_user'] = df['returning_user'].replace([0,1],['New','Returning'])
```

```python
compare_df.rename(columns={'returning_user':'Customer Type'}, inplace=True)
```
